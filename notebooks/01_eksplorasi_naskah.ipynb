# Eksplorasi Naskah Nusantara

Notebook ini digunakan untuk mengeksplorasi dan memproses naskah kuno dari folder `data/raw/`.

```python
# Import library yang dibutuhkan
import sys
from pathlib import Path

# Tambahkan root directory ke path agar bisa import dari src/
sys.path.append(str(Path.cwd().parent))

from src.preprocessing.reader import ManuscriptReader
import matplotlib.pyplot as plt
import pandas as pd
```

## 1. Membaca Naskah

```python
# Inisialisasi reader
reader = ManuscriptReader()

# Baca satu naskah
filename = 'hikayat_sampel_01.txt'
result = reader.process_manuscript(filename)

print(f"File: {result['filename']}")
print(f"Panjang teks: {result['cleaned_length']} karakter")
print("\n" + "="*60)
print(result['cleaned_text'])
```

## 2. Analisis Entitas

```python
# Tampilkan entitas yang terdeteksi
print("Nama yang terdeteksi:")
for name in result['entities']['possible_names']:
    print(f"  - {name}")

print("\nTempat yang terdeteksi:")
for place in result['entities']['possible_places']:
    print(f"  - {place}")
```

## 3. Proses Semua Naskah

```python
import os

# Proses semua file
all_results = []
for filename in os.listdir(reader.data_dir):
    if filename.endswith('.txt'):
        result = reader.process_manuscript(filename)
        all_results.append({
            'filename': filename,
            'length': result['cleaned_length'],
            'num_names': len(result['entities']['possible_names']),
            'num_places': len(result['entities']['possible_places'])
        })

# Buat DataFrame
df = pd.DataFrame(all_results)
print("Ringkasan semua naskah:")
print(df)
```

## 4. Visualisasi

```python
# Visualisasi panjang teks
plt.figure(figsize=(10, 6))
plt.bar(df['filename'], df['length'], color='steelblue')
plt.xlabel('Nama File')
plt.ylabel('Panjang Teks (karakter)')
plt.title('Perbandingan Panjang Naskah')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
